```python
import numpy as np
from matplotlib import pyplot as plt  #matplotlib is package
import random
```

```python
#linspace- equal spaced numbers within a range
x = np.linspace(0,20,num=2000) #there will be 2000 numbers linearly spaced from 0-20
randm = np.random.rand(2000) #gives elements less than 1, this is used to create randomness in data
volts = np.sin(x) + 0.25*randm
current = np.cos(x) + 0.25*randm
charge = np.sin(x)**2 + np.cos(x)+0.25*randm
plt.plot(volts)
plt.plot(current)
plt.plot(charge)
plt.legend("Volts, Current, Charge")
plt.show()
```

```python
np.random.rand(10)
```

```python
ac = {
    "volts" : volts,
    "current" : current,
    "charge" : charge
    }
ac_df = pd.DataFrame(ac)
ac_df.plot()
```
