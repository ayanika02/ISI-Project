```python
for i in symbol_list:
    stock = pd.read_csv(f'{i}.csv', usecols=['Date','Close','Adj Close','Volume'])
    stock.set_index('Date', inplace=True)
    # Rename columns to include the stock symbol
    stock.columns = pd.MultiIndex.from_product([[i], stock.columns])
    #stock['Symbol'] = i
    data.append(stock)
```

```python
returns = pd.DataFrame()
for stock in symbol_list:
    returns[stock+' Returns'] = combined_df[stock]['Adj Close'].pct_change()
```

```python
#single run
np.random.seed(101)
weights = np.array(np.random.random(200))
print('Random weights')
print(weights)
print('Rebalanced weights')
weights = weights/np.sum(weights)
print('Portfolio return')
port_ret = np.sum(returns.mean()*250*weights)
print(port_ret)
print('Expected volatility')
port_vol = np.sqrt(np.dot(weights.T, np.dot(returns.cov()*250,weights)))
print(port_vol)
print('Sharpe Ratio')
SR = (port_ret - 0.075)/port_vol
print(SR)
```

```python
sim = 5000
all_weights = np.zeros((sim, len(symbol_list)))
ret_arr = np.zeros(sim)
vol_arr = np.zeros(sim)
sharpe_arr = np.zeros(sim)

for i in range(sim):
    weights = np.array(np.random.random(200))
    weights = weights/np.sum(weights)
    all_weights[i,:] = weights
    ret_arr[i] = np.sum(returns.mean()*250*weights)
    vol_arr[i] = np.sqrt(np.dot(weights.T,np.dot(returns.cov()*250,weights)))
    sharpe_arr[i] = (ret_arr[i] - 0.075)/vol_arr[i]
```
