```python
sp500 = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
```

```python
sp500
```

```python
sp500['Symbol'] = sp500['Symbol'].str.replace('.','-')
```

```python
symbols = sp500['Symbol'].unique().tolist()
symbols
```

```python
end_date = '2023-12-31'
```

```python
start_date = '2018-01-01'
```

```python
df = yf.download(tickers=symbols,start=start_date,end=end_date)
```

```python
sector = sp500['GICS Sector'].unique()
sector
```

```python
sub_industry = np.array(sp500['GICS Sub-Industry'].unique())
sub_industry
```

```python
types = np.array(["Hydrogen Energy", "Information Technology", "Semiconductor","Housing","Electronic Vehicle","Fin"])
```

```python
for i in range(len(sub_industry)):
    for j in range(len(types)):
        if types[j] in sub_industry[i]:
            print(symbols[i])
```

```python
print(sp500.loc[sp500['Symbol']=='BAC'])
```

```python
it = sp500.loc[sp500['GICS Sector'] == "Information Technology"]
it['GICS Sub-Industry'].unique()
type(it)
```

```python
semiconductors = it.loc[it['GICS Sub-Industry'] == "Semiconductors"]
semiconductors
type(semiconductors)
```

```python
len(it)
```

```python
finance = sp500.loc[sp500['GICS Sector'] == "Financials"]
finance['GICS Sub-Industry'].unique()   
```

```python
keywords = ['fin', 'finance', 'tech','fintech','data']
pattern = '|'.join(keywords)
fintech = sp500[sp500['GICS Sub-Industry'].str.contains(pattern,case=False,na=False)]
fintech
```

```python
len(fintech)
```

```python
energy = sp500.loc[sp500['GICS Sector'] == "Energy"]
energy  #theres no hydrogen
```

```python
keywords = ['property', 'house', 'housing','properties']
pattern = '|'.join(keywords)
housing = sp500[sp500['GICS Sub-Industry'].str.contains(pattern,case=False,na=False)]
housing
```

```python
keywords = ['EV', 'electric', 'vehicle','electric vehicle']
pattern = '|'.join(keywords)
ev = sp500[sp500['GICS Sub-Industry'].str.contains(pattern,case=False,na=False)]
ev
```

```python
dataframes = [ev,housing,energy,fintech,finance,semiconductors,it]
combined_df = pd.concat(dataframes).drop_duplicates().reset_index(drop=True)
```

```python
combined_df
```

```python
combined_df['Symbol'] = combined_df['Symbol'].str.replace('.','-')
symbols = combined_df['Symbol'].unique().tolist()
symbols
```

```python
df = yf.download(tickers=symbols,start=start_date,end=end_date)
```
