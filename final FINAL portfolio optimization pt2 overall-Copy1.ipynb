```python
for i in symbol_list:
    stock = pd.read_csv(f'{i}.csv', usecols=['Date','Close','Adj Close','Volume'])
    stock.set_index('Date', inplace=True)
    # Rename columns to include the stock symbol
    stock.columns = pd.MultiIndex.from_product([[i], stock.columns])
    #stock['Symbol'] = i
    data.append(stock)
```

```python
sim = 5000
all_weights = np.zeros((sim, len(symbol_list)))
ret_arr = np.zeros(sim)
vol_arr = np.zeros(sim)
sharpe_arr = np.zeros(sim)

for i in range(sim):
    weights = np.array(np.random.random(10))
    weights = weights/np.sum(weights)
    all_weights[i,:] = weights
    ret_arr[i] = np.sum(returns.mean()*250*weights)
    vol_arr[i] = np.sqrt(np.dot(weights.T,np.dot(returns.cov()*250,weights)))
    sharpe_arr[i] = (ret_arr[i] - 0.075)/vol_arr[i]
```

```python
df = pd.DataFrame({
    'Symbol': symbol_list,
    'Weight': weights
})

# Sort the DataFrame by the weight values
df_sorted = df.sort_values(by='Weight', ascending=False)

# Display the sorted DataFrame
print(df_sorted)
```

```python
stocks = df_sorted['Symbol'].tolist()
stocks
```

```python
prices = []
start_date = '2020-04-01'
for symbol in stocks:
        # Fetch historical data from Yahoo Finance
        data = yf.download(symbol+".NS", start="2020-04-01", end='2020-04-02')
        price_april2020 = data.loc[start_date]['Adj Close']
        prices.append(price_april2020)
df_prices = pd.DataFrame({'Symbol': stocks, 'Price_April2020': prices})

# Display the prices
print(df_prices)
```

```python
prices=[]
```

```python
start_date = '2024-05-02'
for symbol in stocks:
        # Fetch historical data from Yahoo Finance
        data = yf.download(symbol+".NS", start="2024-05-02", end='2024-05-03')
        price_2024 = data.loc[start_date]['Adj Close']
        prices.append(price_2024)
df_prices2 = pd.DataFrame({'Symbol': stocks, 'Price_2024': prices})

# Display the prices
print(df_prices2)
```

```python
combined_df = pd.merge(df_prices, df_prices2, on='Symbol')
```

```python
combined_df = pd.merge(df_sorted, combined_df, on='Symbol')
```

```python
combined_df
```

```python
portfolio_value = 10000000
combined_df['Investment'] = combined_df['Weight'] * portfolio_value
combined_df['Shares'] = (combined_df['Investment'] / combined_df['Price_April2020']).astype(int)

# Calculate the value of each stock on 1st June 2024
combined_df['Value_2024'] = combined_df['Shares'] * combined_df['Price_2024']

# Calculate total investment used and remaining cash
total_investment_used = (combined_df['Shares'] * combined_df['Price_April2020']).sum()
remaining_cash = portfolio_value - total_investment_used

# Calculate total portfolio value in June 2024 and profit/loss
total_value_2024 = combined_df['Value_2024'].sum()
profit_loss = total_value_2024 - total_investment_used
```

```python
combined_df
```

```python
# Display the final DataFrame and summary
print(combined_df)
print("\nSummary:")
print(f"Total Investment Used: Rs.{total_investment_used:.2f}")
print(f"Remaining Cash: Rs.{remaining_cash:.2f}")
print(f"Total Portfolio Value in June 2024: Rs.{total_value_2024:.2f}")
print(f"Total Profit: Rs.{profit_loss:.2f}")
print(f"Total Profit %: {profit_loss/total_investment_used*100:.2f}")
```
